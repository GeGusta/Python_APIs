### Autenticação

```python
acces_token = 'Seu token de acesso github'
header['Authorization'] = 'Bearer ' + acces_token
```

```python
header
```

```python
response = requests.get(url, headers=header)
response.status_code
```

### Paginação

```python
lista_repo = []

for pg in range(1,7):
    try:
        url_page = f'{url}?page={pg}'
        response = requests.get(url_page, headers=header)
        lista_repo.append(response.json())
    except:
        lista_repo.append(None)
```

```python
repos = 0

for i in lista_repo:
    repos = repos + len(i)

repos
```

```python
url_foll = "https://api.github.com/users/amzn/followers"
```

```python
response = requests.get(url_foll, headers=header)
```

```python
len(response.json())
```

```python
lista_foll = []
tamanho = 1
pg = 1

while tamanho > 0:
    
    try: 
        url_pg = f'{url_foll}?page={pg}'
        response = requests.get(url_pg, headers=header)
        lista_foll.append(response.json())

        tamanho = len(response.json())
        pg = pg + 1
    except: 
        lista_foll.append(None)
```

```python
len(lista_foll)
```

```python
repos = 0

for i in lista_foll:
    repos = repos + len(i)

repos
```

# Tratamento dos Dados


```python
lista_repo[0][0].keys()
```

```python
lista_repo[0][2]["language"]
```

```python
lista_language = []

for pages in lista_repo:
    for repo in pages:
        lista_language.append(repo["language"])
```

```python
lista_name = []

for pages in lista_repo:
    for repo in pages:
        lista_name.append(repo["name"])
```

### Criando Data Frama


```python
Linguagens_Amz = pd.DataFrame()

Linguagens_Amz["Repositorio_Nome"] = lista_name
Linguagens_Amz["Language"] = lista_language

Linguagens_Amz
```

```python
Linguagens_Amz.Language.value_counts()
```

```python
Linguagens_Amz.to_csv("Linguagens_AMZ.csv")
```

## Criando Repositório
O repositório será usado para disponibilizar os dados

```python
url_base = 'https://api.github.com'
url = f'{url_base}/user/repos'

url
```

```python
data = {
    'name': 'linguagens-utilizadas',
    'description': 'Linguagens utilizados nos repositórios da Amazon.',
    'private': False
}
```

```python
response = requests.post(url, json = data, headers=header)
response.status_code
```

### Formato do Arquivo
Para mandar o arquivo para o repositório é necessário que ele esteja em Base64.

```python
with open("Linguagens_AMZ.csv", 'rb') as file:
    conteudo = file.read()

conteudo_codificado = base64.b64encode(conteudo)
```

### PUT nos arquivos

```python
api_base_url = 'https://api.github.com'
username = 'GeGusta'
repo = 'linguagens-utilizadas'
path = "Linguagens_AMZ.csv"

url_put = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'

url_put
```

```python
data = {
    'message': "Adicionar um novo arquivo",
    'content': conteudo_codificado.decode('utf-8')
}
```

```python
response_put = requests.put(url_put,json = data, headers = header)
response_put.status_code
```
